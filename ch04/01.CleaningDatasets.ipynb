```python
url = 'https://raw.githubusercontent.com/kanetkar/LULML/refs/heads/main/ch04/treatments.csv'
treatments = pd.read_csv(url)
print(treatments.info( ))
```

```python
print(treatments.isnull( ).sum( ))
```

```python
print(treatments.duplicated( ).sum( ))
```

```python
print(treatments.describe( ))
```

```python
# fill missing weight_kg with mean weight
mean_weight = patients['weight_kg'].mean( )
patients['weight_kg'] = patients['weight_kg'].fillna(mean_weight)

# fill missing height_cm with mean height
mean_height = patients['height_cm'].mean( )
patients['height_cm'] = patients['height_cm'].fillna(mean_height)

# fill missing contact_info with 'Not Provided'
patients['contact_info'] = patients['contact_info'].fillna('Not Provided')
# verify that there are no missing values
print('Missing weight_kg:', patients['weight_kg'].isnull( ).sum( ))
print('Missing height_cm:', patients['height_cm'].isnull( ).sum( ))
print('Missing contact_info:', patients['contact_info'].isnull( ).sum( ))
```

```python
gender_dict = { 'male': 'Male', 'M': 'Male', 'Male': 'Male',
	       'female': 'Female', 'F': 'Female', 'Female': 'Female' }
patients['gender'] = patients['gender'].map(gender_dict)

# check the unique values in gender
print(patients['gender'].unique( ))
```

```python
state_mapping = {
	                 'Maharashtra': 'Maharashtra',
	                 'MH': 'Maharashtra',
	                 'Karnataka': 'Karnataka',
	                 'KA': 'Karnataka',
	                 'Tamil Nadu': 'Tamil Nadu',
	                 'TN': 'Tamil Nadu',
	                 'West Bengal': 'West Bengal',
  	                 'WB': 'West Bengal',
	                 'Gujarat': 'Gujarat',
	                 'GJ': 'Gujarat'
                                 }
patients['state'] = patients['state'].map(state_mapping)
# check the unique values in state
print(patients['state'].unique( ))
```

```python
# remove special characters from last_name
patients['last_name'] = patients['last_name'].str.replace(r'[^a-zA-Z\s]', '', regex = True)

# check for special characters
sc = patients['last_name'].str.contains(r'[^a-zA-Z\s]').sum( )
print('Special characters in last_name:', sc)
```

```python
# convert birthdate to datetime
patients['birthdate'] = pd.to_datetime(patients['birthdate'],
                                          errors='coerce')
# convert pin_code to string
patients['pin_code'] = patients['pin_code'].astype(str).str.zfill(6)

# check data types
print(patients[['birthdate', 'pin_code']].dtypes)
```

```python
patients = patients.drop_duplicates(subset='patient_id', keep='first')
print(patients.duplicated().sum( ))
```

```python
# cap minimum weight
min_weight = 30  # kg, plausible minimum weight for adults
patients.loc[patients['weight_kg'] < min_weight, 'weight_kg'] = min_weight

# cap maximum height
max_height = 250  # cm, plausible maximum height for humans
patients.loc[patients['height_cm'] > max_height, 'height_cm'] = max_height

# verify that weights are not less than min_weight
print('Minimum weight after capping:', patients['weight_kg'].min( ))

# verify that heights are not greater than max_height
print('Maximum height after capping:', patients['height_cm'].max( ))
```

```python
# split contact_info into email and phone_number
contact_split = patients['contact_info'].str.split(';', expand=True)
patients['email'] = contact_split[0]
patients['phone_number'] = contact_split[1]

# check the new columns
print(patients[['email', 'phone_number']].head( ))
```

```python
patients = patients.drop(columns='contact_info')
```

```python
# melt the dataset to long format
treatments_long = treatments.melt(
    id_vars=['patient_id', 'fasting_blood_sugar_start',
                    'fasting_blood_sugar_end', 'bs_change', 'adverse_event'],
    value_vars=['ayurvedic_medicine_dosage',
                            'allopathic_medicine_dosage'],
    var_name='treatment_type',
    value_name='dosage_range')
print(treatments_long[['treatment_type', 'dosage_range']].head( ))
```

```python
treatments_long['treatment_type'] = treatments_long['treatment_type'].str.replace('_medicine_dosage', '').str.capitalize( )
print(treatments_long[['treatment_type', 'dosage_range']].head( ))
```

```python
# remove records where dosage_range is '-'
treatments_long = treatments_long[treatments_long['dosage_range'] != '-']
print(treatments_long[['treatment_type', 'dosage_range']].head( ))
print(treatments_long.shape)
```

```python
# split dosage_range into dosage_start and dosage_end
treatments_long['dosage_start'] = treatments_long['dosage_range'].str.split('-').str.get(0)
treatments_long['dosage_end'] = treatments_long['dosage_range'].str.split('-').str.get(1)

# view the new dosage columns
print(treatments_long[['dosage_start', 'dosage_end']].head( ))
```

```python
# drop the dosage_range column
treatments_long.drop(columns='dosage_range', inplace=True)
```

```python
# remove 'mg' from dosage_start and dosage_end
treatments_long['dosage_start'] = treatments_long['dosage_start'].str.replace('mg', '').str.strip( )
treatments_long['dosage_end'] = treatments_long['dosage_end'].str.replace('mg', '').str.strip( )
print(treatments_long[['dosage_start', 'dosage_end']].head( ))
```

```python
# convert dosage columns to numeric
treatments_long['dosage_start'] = pd.to_numeric(treatments_long['dosage_start'], errors='coerce')
treatments_long['dosage_end'] = pd.to_numeric(treatments_long['dosage_end'], errors='coerce')

# verify data types
print(treatments_long[['dosage_start', 'dosage_end']].dtypes)
```

```python
# calculate bs_change if missing
treatments_long['bs_change'] = treatments_long['fasting_blood_sugar_end'] - treatments_long['fasting_blood_sugar_start']

# fill missing adverse_event with 'None'
treatments_long['adverse_event'] = treatments_long['adverse_event'].fillna('None')

# check for missing values
print('Missing bs_change:', treatments_long['bs_change'].isnull( ).sum( ))
print('Missing adverse_event:', treatments_long['adverse_event'].isnull( ).sum( ))
```

```python
# save the cleaned dataset
patients.to_csv('patients_cleaned.csv', index = False)
treatments_long.to_csv('treatments_cleaned.csv', index = False)
```
