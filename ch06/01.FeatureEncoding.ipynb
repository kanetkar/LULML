```python
import pandas as pd

url = 'https://raw.githubusercontent.com/kanetkar/LULML/refs/heads/main/ch06/customers.csv'
df = pd.read_csv(url)
print(df.head( ))
```

```python
from sklearn.preprocessing import OrdinalEncoder

# define the columns to encode and their order
ordinal_columns = ['review', 'education']
ordinal_categories = [
						['Poor', 'Average', 'Good'],
						['School', 'UG', 'PG']
					   ]

# initialize the OrdinalEncoder with the specified category orders
oe = OrdinalEncoder(categories = ordinal_categories)

# apply the encoding
df[ordinal_columns] = oe.fit_transform(df[ordinal_columns])
print(df.head( ))
```

```python
from sklearn.preprocessing import OneHotEncoder

# initialize the OneHotEncoder
ohe = OneHotEncoder(sparse_output = False, drop = None)

# fit and transform the gender column
gender_encoded = ohe.fit_transform(df[['gender']])

# convert dense array to a data frame
gender_encoded_df = pd.DataFrame(gender_encoded, columns = ohe.get_feature_names_out(['gender']))

# concatenate the new columns to the original dataset
df = pd.concat([df, gender_encoded_df], axis = 1)

# drop the original gender column
df.drop('gender', axis = 1, inplace = True)

# display the first few rows of the updated dataset
print(df[['age', 'gender_Female', 'gender_Male']].head( ))
#print(df.head( ))

```

```python
from sklearn.preprocessing import LabelEncoder

# initialize the LabelEncoder
label_encoder = LabelEncoder( )

# fit and transform the purchased column
df['purchased'] = label_encoder.fit_transform(df['purchased'])

# display the first few rows of the updated dataset
print(df[['age', 'purchased']].head( ))
```
