```python
import pandas as pd
url = 'https://raw.githubusercontent.com/kanetkar/LULML/refs/heads/main/ch07/finance.csv'
df = pd.read_csv(url)
print(df.sample(5))
```

```python
print((df.isnull( ).sum( ) / df.shape[0]) * 100)
```

```python
import seaborn as sns
import matplotlib.pyplot as plt

new_df = df[df['Credit_Score'].notnull( )]
sns.histplot(df['Age'], kde=True, label='before deletion')
sns.histplot(new_df['Age'], kde=True , label='after deletion')
plt.legend( )
plt.show( )
```

```python
new_df = df[df['Credit_Score'].notnull( )]
sns.histplot(df['Monthly_Expenses'], kde=True, label='before del')
sns.histplot(new_df['Monthly_Expenses'], kde=True, label='after del')
plt.legend( )
plt.show( )
```

```python
import matplotlib.pyplot as plt

new_df = df[df['Credit_Score'].notnull( )]
cnt_before = df['Loan_Type'].value_counts(normalize=True)
cnt_after = new_df['Loan_Type'].value_counts(normalize=True)

# combine the two series into a single DataFrame
combo = pd.DataFrame({'Before Del': cnt_before,
                                             'After Del': cnt_after})

combo.plot(kind='bar', color=['#FFB300', '#F26B21'], figsize=(8, 4))
plt.title('Distribution of Loan Type (Before and After Deletion)')
plt.xlabel('Loan Type')
plt.ylabel('Proportion')
plt.xticks(rotation=0)
plt.grid(axis='y', linestyle=':', alpha=0.7)
plt.legend( )
plt.tight_layout( )
plt.show( )

```
