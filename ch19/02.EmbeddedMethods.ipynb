```python
from sklearn.datasets import load_breast_cancer
from sklearn.model_selection import train_test_split
from sklearn.ensemble import RandomForestClassifier

# load dataset
X, y = load_breast_cancer(return_X_y = True)

# train-test split
X_train, X_test, y_train, y_test = train_test_split(X, y,
	test_size = 0.2, random_state = 1, stratify = y)

# train Random Forest
rf = RandomForestClassifier(n_estimators = 300, random_state = 1)
rf.fit(X_train, y_train)
```

```python
import pandas as pd
from sklearn.datasets import load_breast_cancer
dataset = load_breast_cancer( )

feature_importances = rf.feature_importances_
feature_names = dataset.feature_names

importance_df = pd.DataFrame({"feature": feature_names,
				      "importance": feature_importances}).sort_values(by = "importance", ascending = False)

print(importance_df.head(10))
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize = (12,6))
bars = plt.bar(
    importance_df["feature"],
    importance_df["importance"],
    color = "steelblue",
    label = "importance")

plt.title("Top 10 Feature Importances", fontsize = 14)
plt.xlabel("Feature Name", fontsize = 12)
plt.ylabel("Importance", fontsize = 12)
plt.xticks(rotation = 45, ha = "right")
plt.legend( )
plt.tight_layout( )
plt.show( )
```
